```python
#随机漫步：随机漫步是一种数学统计模型，它由一连串轨迹所组成，其中每一次都是随机的，它能用来表示不规则的变动形式，如同一个人乱步所形成的随机记录。
import random
import matplotlib.pyplot as plt
position=0
walk=[position]
steps=1000    # 实验次数
for i in range(steps):
    step=1 if random.randint(0,1) else -1   # 0,1取随机整数，0则step前进1，1则后退1布，有点像抛硬币
    position +=step                          # position位置前后移动
    walk.append(position)
#plt.plot(walk[:100])    #切片
# 需求：一次性抽取1000个投币的结果，每次投币的结果为1或者-1，然后计算累计值
draws=np.random.choice((-1,1),size=1000)  #参数：low、high、size；取值范围：[low,hight)
plt.plot(draws[:100])
walk=draws.cumsum()
#plt.plot(walk[:100])
# 从漫步轨道上取最值
walk.min()
walk.max()
#argmax（）或者argmax（0）：求数组中最大值的第一个位置
(np.abs(walk)>=10).argmax(0)   #求第一次累加超过10或者-10的数组的位置  
#但是上述方法有一个问题就是：不是每一个数组都会有超过10或者-10，或者其他标量的位置
#解决办法：用any来判断是否满足条件,存在就返回true，否则返回false---多次随机漫步
#多次随机漫步
nwalks=5000
nsteps=1000
walks=(np.random.choice((-1,1),size=(nwalks,nsteps))).cumsum(1)

hit30=(np.abs(walks)>=30).any(1)
plt.plot(walk[:])
hit30.sum()   #漫步5000步中，超过30 的漫步次数
time30=(np.abs(walks[hit30])>=30).argmax(1)  #在超过30的数组里面，第一次步数和超过30的步数
time30.mean()
```

```python
import numpy as np
import pandas as pd
a=[100,400,600,800,3000,4200]
np.mean(a)
sum=0
for i in a:
    sum=(i-100)*(100-0)/(4200-100)
#     print(sum)

age=[8,10, 15, 35, 50, 52, 85, 89, 92, 158, 201, 251]
# qcut函数
# (1). 基于分位数（几分中位数）进行切割
cats=pd.cut(age,3) # 箱体分别为：(-0.001, 7.5], (7.5, 15.0], (15.0, 22.5], (b22.5, 32.0]
cats.codes  # 识别并标记age中的数据在哪个箱子，没有分到箱子的数据标记为-1
# cats.categories # 显示目前所用的箱体：(10, 20], (20, 30]
# cats.value_counts(cats)  # 统计每个箱体中的含有数据数量


# (2).按照比例进行划分：[比例大小]
# print(pd.qcut(age,[0,0.25,1],labels=['small number','large number']))
#每个箱体各占箱子长度的50%，对半分，并给箱体贴上name：['small number','large number']
# print(pd.qcut(age,[0, 0.1, 0.2, 0.3, 1],labels=['first 10%','second 10%','third 10%','70%']))
#每个箱体各占箱子长度的10，210，10，70%，对半分，并给箱体贴上name：['first 10%','second 10%','third 10%','70%']

df=pd.DataFrame(
        data={
            'X':[0,0,3],
            'Y':[0,2,4]},index=['P1','P2','P3'])
df['X'].iloc[2]
# def mink(df,r):
#     i=0
#     sum=0
#     for i,j in zip(df['X'],df['Y']):
#         for m,n in zip(df['X'],df['Y']):
#             sum=np.abs(i-m)**r+np.abs(j-n)**r
#             sum=np.power(sum,1/r)
#             print(i,' ',j,' ',m,' ',n,'=',sum)
# mink(df,np.inf)
df=pd.DataFrame(
        data={
            'X':[1,0,1,0,1,0,1,0],
            'Y':[1,1,0,0,1,1,0,0],
            'Z':[1]*4+[0]*4,
            'class':[1*5,-1*20,1*20,-1*5,0,1*25,0,-1*25]})

df=df[df['X']==0]

# 属性
def cal(c,r):
    dfxt=df[df[c]==r]
    xt=dfxt['class'].abs().sum()
# + 个数
    dfx1=df[(df[c]==r)&(df['class']>0)]
    x1=np.abs(dfx1['class'].sum())
# - 个数
    dfx0=df[(df[c]==r)&(df['class']<0)]
    x0=np.abs(dfx0['class'].sum())
# info
    infoxt=1-(x1/xt)**2-(x0/xt)**2
    print(x1,' ',x0,'   ',xt,'   ',infoxt)#info
    return infoxt

def m(c):
    df1=df[df[c]==1]
    num1=float(df1['class'].abs().sum())
    df2=df[df[c]==0]
    num2=float(df2['class'].abs().sum())
    print(num1,'  ',num2)
    result=(num1/(num1+num2))*cal(c,1)+(num2/(num1+num2))*cal(c,0)#info(x,y)  
    print(num1/(num1+num2),'   ',num2/(num1+num2),'  ',result)
    print(0.5-result)#gain
m('Y')
df[df['Y']==1]
```
